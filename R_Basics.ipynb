## Next statement

```r
# with the next stats we can skip an iteration
i<-0
while(i<10){
  i<-i+1
  if(i==3){
    next
  }
  print(i)
}
```

```r
#YAhteez
# if_else condition with a while loop to demostrated a practical exaple let us sya we play a game of yahteez!
dice<-i
while(dice<=6){
  if(dice<6){
    print("no yahteez!")
  }
  dice<-dice+1
  print("yahteez")
}
```

## for loop

```r
for (x in 1:10){
  print(x)
}
```

```r
# nested loop
# print the adjective of the each element in a list
adj<-list("red","big","testy","sweet")
fruits<-list("apple","banana","cherry","orange")
for (x in  adj){
  for (y in fruits){
    print(paste(x,y))
  }
}
```
